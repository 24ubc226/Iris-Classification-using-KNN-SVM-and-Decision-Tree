# ML: TASK-03 Practical Study of Machine Learning Classifiers on Iris Species Data

## 1. Load and Preprocess Dataset

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the Iris dataset

```python
# Load the Iris dataset
iris_df = pd.read_csv('/content/Iris.csv')

# Display the first 5 rows to understand the data structure
display(iris_df.head())
```

### Initial Data Exploration and Preprocessing

```python
# Check for missing values
print("\nMissing values:\n", iris_df.isnull().sum())

# Get a summary of the dataframe
print("\nDataFrame Info:\n")
iris_df.info()

# Get descriptive statistics
print("\nDescriptive Statistics:\n")
display(iris_df.describe())
```

```python
# Drop the 'Id' column as it's not a feature for classification
if 'Id' in iris_df.columns:
    iris_df = iris_df.drop('Id', axis=1)

# Separate features (X) and target (y)
X = iris_df.drop('Species', axis=1)
y = iris_df['Species']

# Encode the target variable 'Species' into numerical format
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

## 2. Implement Models

### K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize and train the KNN model
knn_model = KNeighborsClassifier(n_neighbors=5) # Using 5 neighbors as a common starting point
knn_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_knn = knn_model.predict(X_test)

# Evaluate KNN model
accuracy_knn = accuracy_score(y_test, y_pred_knn)
conf_matrix_knn = confusion_matrix(y_test, y_pred_knn)

print(f"KNN Accuracy: {accuracy_knn:.4f}")
print("KNN Confusion Matrix:\n", conf_matrix_knn)
```

### Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

svm_results = {}
kernels = ['linear', 'rbf', 'sigmoid', 'poly']

for kernel_type in kernels:
    print(f"\nTraining SVM with {kernel_type} kernel...")
    svm_model = SVC(kernel=kernel_type, random_state=42)
    svm_model.fit(X_train, y_train)
    y_pred_svm = svm_model.predict(X_test)

    accuracy_svm = accuracy_score(y_test, y_pred_svm)
    conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)

    svm_results[kernel_type] = {
        'accuracy': accuracy_svm,
        'confusion_matrix': conf_matrix_svm
    }

    print(f"SVM ({kernel_type} kernel) Accuracy: {accuracy_svm:.4f}")
    print(f"SVM ({kernel_type} kernel) Confusion Matrix:\n", conf_matrix_svm)
```

## 3. Implement Additional Algorithm: Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train the Decision Tree model
decision_tree_model = DecisionTreeClassifier(random_state=42)
decision_tree_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_dt = decision_tree_model.predict(X_test)

# Evaluate Decision Tree model
accuracy_dt = accuracy_score(y_test, y_pred_dt)
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)

print(f"Decision Tree Accuracy: {accuracy_dt:.4f}")
print("Decision Tree Confusion Matrix:\n", conf_matrix_dt)
```

## 4. Model Evaluation & 5. Model Comparison

### Summary of Model Performance

```python
print("\n--- Model Performance Summary ---")
print(f"K-Nearest Neighbors (KNN) Accuracy: {accuracy_knn:.4f}")
print(f"Decision Tree Accuracy: {accuracy_dt:.4f}")

print("\nSupport Vector Machine (SVM) Accuracies:")
for kernel, results in svm_results.items():
    print(f"- {kernel.capitalize()} kernel: {results['accuracy']:.4f}")

# You can also display confusion matrices for comparison if needed
# print("\nKNN Confusion Matrix:\n", conf_matrix_knn)
# print("\nDecision Tree Confusion Matrix:\n", conf_matrix_dt)
# for kernel, results in svm_results.items():
#     print(f"\nSVM ({kernel.capitalize()} kernel) Confusion Matrix:\n", results['confusion_matrix'])
```

# Research & Understanding Component:
## Decision Tree
Conceptually‚ how the Decision Tree algorithm works
Let's have some more detail on how the Decision Tree works․ The Decision Tree is a powerful and versatile non-parametric supervised learning algorithm capable of performing classification and regression tasks․ It is constructed as a flow chart‚ in which a number of questions depending on the data are asked․ The internal 'node' of the tree will be a 'test' on a feature‚ for example‚ 'Is SepalLengthCm greater than 5․0?'․ The 'branch' coming out of the test is based on the outcome of the test․ You then reach a 'leaf node' where the actual outcome is determined‚ either the class label (for classification) or a numerical value (for regression)․.

The goal is essentially to predict a target variable by learning straightforward decision rules directly from your data's features. The algorithm cleverly figures out the 'best' feature to split on at each step by evaluating metrics like Gini impurity or entropy. This splitting process continues until a stop condition is met, such as when all data points in a subset belong to the same class, there are no more features left to split on, or a predefined maximum depth for the tree is reached.

### Its key assumptions

While Decision Trees are quite flexible, they do operate with a few underlying ideas:

1.  **Feature Importance:** The algorithm implicitly assumes that some features are more crucial than others for predicting the target. These more important features tend to be used for splits higher up in the tree.
2.  **Greedy Approach:** It takes a 'greedy' approach: at each step, it makes the best local decision (the split that maximizes information gain or minimizes impurity) hoping this leads to a strong overall tree. It doesn't look back or re-evaluate past decisions, which means it might not always find the absolute best *global* tree.
3.  **No Data Distribution Assumptions:** One of its great strengths is that, unlike some other algorithms (think Naive Bayes or Linear Regression), Decision Trees don't make any specific assumptions about how your data is distributed. This makes them very adaptable!

### Strengths and limitations

Like any tool, Decision Trees have their superpowers and their Achilles' heels:

**Strengths (Why we love them!):**

*   **Easy to Understand and Interpret:** They're incredibly intuitive! You can actually visualize a decision tree and easily follow its logic, making its predictions very transparent. This is fantastic for explaining models to non-technical folks.
*   **Handles Both Numerical and Categorical Data:** You don't need to do extensive preprocessing like one-hot encoding for categorical features; Decision Trees can work with both types of data quite well.
*   **No Feature Scaling Required:** Forget about standardizing or normalizing your features! Decision Trees aren't sensitive to the scale of your data.
*   **Captures Non-linear Relationships:** They're great at modeling complex, non-linear relationships between your features and the target variable, which many other algorithms struggle with.

**Limitations (Where they can stumble):**

*   **Prone to Overfitting:** One big challenge is overfitting: without careful 'pruning' or setting limits (like a `max_depth`), a single Decision Tree can easily become too complex, memorizing the training data rather than learning general patterns. This leads to poor performance on new, unseen data.
*   **Instability:** Small changes in your training data can sometimes lead to a completely different tree structure, making them a bit unstable.
*   **Bias towards Dominant Classes:** If your dataset has imbalanced classes, Decision Trees might get a bit biased towards the more frequent classes.
*   **Sub-optimality of Greedy Approach:** As mentioned, because they're greedy, they might not always find the *absolute best* possible tree, even if they find a very good one.

### Situations where it performs well

Decision Trees are a fantastic choice when:

*   **Interpretability is Crucial:** In fields like medicine, finance, or legal, understanding *why* a decision was made is often as important as the decision itself. Their transparency shines here.
*   **Exploring Data:** They can be a great tool for initial exploratory data analysis, helping you quickly understand feature relationships.
*   **Datasets with Mixed Data Types:** If your dataset has a blend of numerical and categorical information, Decision Trees handle this gracefully.
*   **Non-linear Decision Boundaries:** When the relationship between your features and target is complex and doesn't follow a straight line, they can capture those nuances effectively.

### Situations where it may not be suitable

However, you might want to consider other options if:

*   **Extremely Complex Relationships:** For highly intricate relationships, a single Decision Tree might become excessively deep and overly complex, potentially overfitting or simply not capturing all the subtleties.
*   **High-Dimensional Data with Sparse Features:** While they can manage many features, extremely high-dimensional datasets with lots of sparse (mostly zero) features can be challenging.
*   **When Ensemble Methods are Better Suited:** Often, more sophisticated techniques like Random Forests or Gradient Boosting (which are actually built upon many Decision Trees!) can outperform a single Decision Tree. These ensemble methods typically mitigate the limitations of individual trees (like overfitting and instability) at the cost of some interpretability.
